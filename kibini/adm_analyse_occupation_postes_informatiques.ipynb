```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

import datetime

from kiblib.utils.db import DbConn
import warnings
```

```python
warnings('ignore')
```

```python
db_conn = DbConn().create_engine()
```

```python
query = "SELECT * FROM statdb.stat_webkiosk WHERE YEAR(heure_deb) >= 2022" 
df = pd.read_sql(query, con=db_conn)
```

```python
df['heure_deb'] = pd.to_datetime(df['heure_deb'])
df['heure_fin'] = pd.to_datetime(df['heure_fin'])
```

```python
df
```

```python
df
```

```python
connexions = pd.DataFrame()
start_date = datetime.datetime(2024,1,1)
end_date = datetime.datetime(2024,7,20)
delta = datetime.timedelta(days=1)
while (start_date <= end_date):
    for hour in range(9, 20):
        for minute in [0, 15, 30, 45]:
            date_heure = start_date.replace(hour=hour, minute=minute)
            c = df[(df['heure_deb'] <= date_heure)
                & (df['heure_fin'] > date_heure)]
            c['date_heure_connexion'] = date_heure
            connexions = pd.concat([connexions, c])
    start_date = start_date + delta
connexions['nb'] = 1
```

```python
connexions
```

```python
connexions.groupby('espace')['nb'].sum()
```

```python
postes_connectes = connexions.groupby(['date_heure_connexion', 'espace'])['nb'].sum()
postes_connectes = postes_connectes.reset_index()
#joursem = {0: 'lundi', 1: 'mardi', 2: 'mercredi', 3: 'jeudi',
#           4: 'vendredi', 5: 'samedi', 6: 'dimanche'}
postes_connectes['joursem'] = postes_connectes['date_heure_connexion'].dt.dayofweek#.map(joursem)
postes_connectes['semaine'] = postes_connectes['date_heure_connexion'].dt.week
postes_connectes['annee'] = postes_connectes['date_heure_connexion'].dt.year
postes_connectes['heure'] = postes_connectes['date_heure_connexion'].dt.time #strftime('%H:%M')

postes_connectes['periode'] = 'scolaire'
postes_connectes.loc[(postes_connectes['annee'] == 2022) &
                     (postes_connectes['semaine'].isin([43,44,51,52])), 'periode'] = 'vacances'
postes_connectes.loc[(postes_connectes['annee'] == 2023) &
                     (postes_connectes['semaine'].isin([7,8])), 'periode'] = 'vacances'

postes_connectes = postes_connectes[postes_connectes['heure'] <= datetime.time(19)]
postes_connectes = postes_connectes[postes_connectes['heure'] >= datetime.time(9)]
postes_connectes = postes_connectes[~( (postes_connectes['heure'] > datetime.time(18))
                                      & (postes_connectes['periode'] == 'vacances' ) )]
```

```python
postes_connectes_espaces = pd.pivot_table(postes_connectes, values='nb', index=['date_heure_connexion' , 'joursem', 'heure', 'periode'],
                    columns=['espace'], aggfunc=np.sum, fill_value=0, margins=False, margins_name='Total').reset_index()
#postes_connectes_espaces
```

```python
postes_connectes_espaces['Rdc_disp'] = 12
postes_connectes_espaces.loc[postes_connectes_espaces['Rdc'] > postes_connectes_espaces['Rdc_disp'], 'Rdc'] = postes_connectes_espaces['Rdc_disp']
postes_connectes_espaces['tx_rdc'] = round( postes_connectes_espaces['Rdc'] / postes_connectes_espaces['Rdc_disp'],1)

#postes_connectes_espaces['Cafe_disp'] = 3
#postes_connectes_espaces.loc[postes_connectes_espaces['Cafe'] > postes_connectes_espaces['Cafe_disp'], 'Cafe'] = postes_connectes_espaces['Cafe_disp']
#postes_connectes_espaces['tx_cafe'] = round( postes_connectes_espaces['Cafe'] / postes_connectes_espaces['Cafe_disp'],1)

#postes_connectes_espaces['Rdc_Cafe_disp'] = postes_connectes_espaces['Rdc_disp'] + postes_connectes_espaces['Cafe_disp'] 
#postes_connectes_espaces['Rdc_Cafe'] = postes_connectes_espaces['Rdc'] + postes_connectes_espaces['Cafe']
#postes_connectes_espaces['tx_rdc_cafe'] = round( postes_connectes_espaces['Rdc_Cafe'] / postes_connectes_espaces['Rdc_Cafe_disp'],1)

postes_connectes_espaces['Etude_disp'] = 4
postes_connectes_espaces.loc[postes_connectes_espaces['heure'] < datetime.time(10, 0, 0), 'Etude_disp'] = 0
postes_connectes_espaces.loc[postes_connectes_espaces['Etude'] > postes_connectes_espaces['Etude_disp'], 'Etude'] = postes_connectes_espaces['Etude_disp']
postes_connectes_espaces['tx_etude'] = round( postes_connectes_espaces['Etude'] / postes_connectes_espaces['Etude_disp'],1)

postes_connectes_espaces['Jeunesse_disp'] = 4
postes_connectes_espaces.loc[postes_connectes_espaces['heure'] < datetime.time(10, 0, 0), 'Jeunesse_disp'] = 0
postes_connectes_espaces.loc[postes_connectes_espaces['Jeunesse'] > postes_connectes_espaces['Jeunesse_disp'], 'Jeunesse'] = postes_connectes_espaces['Jeunesse_disp']
postes_connectes_espaces['tx_Jeunesse'] = round( postes_connectes_espaces['Jeunesse'] / postes_connectes_espaces['Jeunesse_disp'],1)

postes_connectes_espaces['LAC_disp'] = 3
postes_connectes_espaces.loc[postes_connectes_espaces['heure'] < datetime.time(10, 0, 0), 'LAC_disp'] = 0
postes_connectes_espaces.loc[postes_connectes_espaces['LAC'] > postes_connectes_espaces['LAC_disp'], 'LAC'] = postes_connectes_espaces['LAC_disp']
postes_connectes_espaces['tx_LAC'] = round( postes_connectes_espaces['LAC'] / postes_connectes_espaces['LAC_disp'],1)

postes_connectes_espaces['Phare_disp'] = 4
postes_connectes_espaces.loc[postes_connectes_espaces['heure'] < datetime.time(10, 0, 0), 'Phare_disp'] = 0
postes_connectes_espaces.loc[postes_connectes_espaces['Phare'] > postes_connectes_espaces['Phare_disp'], 'Phare'] = postes_connectes_espaces['Phare_disp']
postes_connectes_espaces['tx_Phare'] = round( postes_connectes_espaces['Phare'] / postes_connectes_espaces['Phare_disp'],1)

postes_connectes_espaces['Xia_disp'] = 0
postes_connectes_espaces.loc[(postes_connectes_espaces['joursem'].isin([2, 5]))
          & (postes_connectes_espaces['heure'] >= datetime.time(14, 0, 0)) , 'Xia_disp'] = 16
postes_connectes_espaces.loc[postes_connectes_espaces['Xia'] > postes_connectes_espaces['Xia_disp'], 'Xia'] = postes_connectes_espaces['Xia_disp']
postes_connectes_espaces['tx_Xia'] = round( postes_connectes_espaces['Xia'] / postes_connectes_espaces['Xia_disp'],1)

postes_connectes_espaces['total_utilise'] = postes_connectes_espaces['Xia'] +  postes_connectes_espaces['Cafe'] + postes_connectes_espaces['Etude'] + postes_connectes_espaces['Jeunesse'] + postes_connectes_espaces['LAC'] + postes_connectes_espaces['Rdc'] + postes_connectes_espaces['Phare']
postes_connectes_espaces['total_disp'] = postes_connectes_espaces['Xia_disp'] +  postes_connectes_espaces['Cafe_disp'] + postes_connectes_espaces['Etude_disp'] + postes_connectes_espaces['Jeunesse_disp'] + postes_connectes_espaces['LAC_disp'] + postes_connectes_espaces['Rdc_disp'] + postes_connectes_espaces['Phare_disp']
postes_connectes_espaces['tx_utilisation'] = round( postes_connectes_espaces['total_utilise'] / postes_connectes_espaces['total_disp'],1)

postes_connectes_espaces = postes_connectes_espaces.fillna(0.0)

#postes_connectes_espaces
```

```python
postes_connectes_espaces_short = postes_connectes_espaces[['date_heure_connexion', 'joursem', 'heure', 'periode', 'tx_rdc', 'tx_cafe', 'tx_rdc_cafe', 'tx_etude', 'tx_jeux', 'tx_lecture',
       'tx_reussir', 'tx_atelier', 'tx_utilisation']]
```

```python
for month in [9, 10, 11, 12, 1, 2, 3, 4]:
    for periode in ['scolaire', 'vacances']:
        for espace in ['tx_rdc_cafe', 'tx_etude', 'tx_jeux', 'tx_lecture', 'tx_reussir', 'tx_atelier']:
            data = postes_connectes_espaces_short[(postes_connectes_espaces_short['periode'] == periode)
                                                 & (postes_connectes_espaces_short['date_heure_connexion'].dt.month == month)]
            if len(data) > 0:
                data_pivot = pd.pivot_table(data, values=espace, index=['heure'],
                                columns=['joursem'], aggfunc=np.median)
                f, ax = plt.subplots(figsize=(15, 10))
                sns.heatmap(data_pivot, cmap='rocket_r', annot=True, fmt='g', ax=ax)
                ax.set_title(f"Mois {month} - pÃ©riode {periode} - {espace}")
                f.tight_layout()
```

```python
weeks = [*range(32,52)] + [*range(1,17)]
for week in weeks:
    for espace in ['tx_rdc_cafe', 'tx_etude', 'tx_jeux', 'tx_lecture', 'tx_reussir', 'tx_atelier']:
        data = postes_connectes_espaces_short[(postes_connectes_espaces_short['date_heure_connexion'].dt.week == week)]
        if len(data) > 0:
            data_pivot = pd.pivot_table(data, values=espace, index=['heure'],
                                columns=['joursem'], aggfunc=np.median)
            f, ax = plt.subplots(figsize=(15, 10))
            sns.heatmap(data_pivot, cmap='rocket_r', annot=True, fmt='g', ax=ax)
            ax.set_title(f"Semaine {week} - {espace}")
            #ax.set_title(f"Semaine {week}")
            f.tight_layout()
```

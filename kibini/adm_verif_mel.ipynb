```python
import pandas as pd
import numpy as np

from datetime import datetime

import requests
import json
```

```python
adherents = pd.read_json("http://cataloguekoha.ntrbx.local/cgi-bin/koha/rest.pl/user")
adherents = adherents[['cardnumber', 'dateofbirth', 'dateexpiry']]
adherents['cardnumber'] = adherents['cardnumber'].str.lower()
adherents['dateexpiry'] = pd.to_datetime(adherents['dateexpiry'])
adherents['dateofbirth'] = pd.to_datetime(adherents['dateofbirth'])
adherents = adherents[adherents['dateexpiry'] >= datetime.now()]
```

```python
demandes = pd.read_csv("/home/kibini/validation_99_2024_06_18.csv")
demandes = demandes[['NÂ° de carte', 'Date de naissance (format jj/mm/aaaa)']]
demandes.columns = ['cardnumber', 'demande_dateofbirth']
demandes['cardnumber'] = demandes['cardnumber'].str.lower()
```

```python
demandes['demande_dateofbirth'] = pd.to_datetime(demandes['demande_dateofbirth'], errors='coerce', dayfirst=True)
```

```python
adherents['statut'] = True
demandes_statut = demandes.merge(adherents, how='left', on='cardnumber')
demandes_statut.loc[demandes_statut['statut'].isna() , 'statut'] = False
demandes_statut['ddn_statut'] = False
demandes_statut.loc[demandes_statut['statut'] == True, 'ddn_statut'] = True
demandes_statut.loc[demandes_statut['dateofbirth'] != demandes_statut['demande_dateofbirth'], 'ddn_statut'] = False
```

```python
def get_all_items():
    items = []
     
    url = f"{base_url}/api/v1/items?_per_page=2000"
    print(url)
    response = requests.get(url, auth=(username, password))
    items.extend(response.json())

    while 'next' in response.links:
        url = response.links['next']['url']
        print(url)
        response = requests.get(url, auth=(username, password))
        items.extend(response.json())
    return items
```

```python
# ajout des colonnes pour aggrégation

prints['date_impression'] = pd.DatetimeIndex(prints['date_impression']).date
prints['annee'] =  pd.DatetimeIndex(prints['date_impression']).year
prints
```

```python
repartition_impressions = prints.groupby(['date_impression','id','annee'])['nb_pages_imprimees'].sum().reset_index()
```

```python
repartition_impressions = repartition_impressions.groupby(['annee','nb_pages_imprimees']).size().reset_index().rename(columns={0:'nb_adherents'})
repartition_impressions
```

```python
repartition_impressions['nb_pages_imprimees_total'] = repartition_impressions['nb_pages_imprimees'] * repartition_impressions['nb_adherents']
```

```python
# V1 : fonction qui sort pour chaque année entrée en paramètre, la part des pages_imprimées & part d'adherents

def RepartitionImpressions(df,annee):
    resultat = df[df['annee']==annee]
    resultat['part_pages_imprimees'] = ((resultat['nb_pages_imprimees_total'] / resultat['nb_pages_imprimees_total'].sum())*100).round(2)
    resultat['part_adherents'] = ((resultat['nb_adherents'] / resultat['nb_adherents'].sum())*100).round(2)
    resultat = pd.pivot_table(data=resultat,index='nb_pages_imprimees',values=['part_pages_imprimees','nb_pages_imprimees_total','part_adherents','nb_adherents'])
    resultat = resultat.loc[:, ['part_pages_imprimees','nb_pages_imprimees_total','part_adherents','nb_adherents']]
    resultat = resultat.sort_values(by=['part_pages_imprimees'],ascending=False)
    resultat = resultat.head(11)
    return(resultat)
```

# V2 Résultat avec AJOUT DE L'INFO TOTAL ADHERENTS ET TOTAL PAGES IMPRIMÉES sur toute l'année

def RepartitionImpressions(df,annee):
    resultat = df[df['annee']==annee]
    resultat['part_pages_imprimees'] = ((resultat['nb_pages_imprimees_total'] / resultat['nb_pages_imprimees_total'].sum())*100).round(2)
    resultat['part_adherents'] = ((resultat['nb_adherents'] / resultat['nb_adherents'].sum())*100).round(2)
    resultat['total_pages_imprimees'] = resultat['nb_pages_imprimees_total'].sum()
    resultat['total_adherents'] = resultat['nb_adherents'].sum()
    resultat = pd.pivot_table(data=resultat,index='nb_pages_imprimees',values=['nb_pages_imprimees_total','part_pages_imprimees','total_pages_imprimees','nb_adherents','part_adherents','total_adherents'])
    resultat = resultat.sort_values(by=['part_pages_imprimees'],ascending=False)
    resultat = resultat.head(11)
    return(resultat)

# Explicitation des colonnes : 

* `nb_pages_imprimees` : il faut le prendre comme un taux multiplcateur. Taux de pages imprimées
* `nb_adherents` : Nombre d'adhérents qui ont lancé une impression au moins une fois dans la journée
* `nb_pages_imprimees_total` : Nombre total de pages imprimées au total (= `nb_pages_imprimees` multiplié par le `nb_adherents`
